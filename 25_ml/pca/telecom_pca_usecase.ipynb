# PCA Use Case using Telecom Customer Dataset

This notebook demonstrates **Principal Component Analysis (PCA)** using a telecom customer dataset.

PCA is an **unsupervised dimensionality reduction technique**.

It is used to reduce many input columns into fewer important components while keeping most of the useful information.

## Telecom use case

A telecom company has many customer behavior columns such as:

- monthly charges
- total charges
- tenure
- data usage
- call minutes
- SMS count
- service calls
- payment delay

Instead of using many columns directly, PCA helps us convert them into fewer principal components.

These components can be used for:

- customer segmentation
- visualization
- noise reduction
- reducing model complexity
- preprocessing before clustering or machine learning


## 1. Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

## 2. Load the Telecom Dataset

```python
# Load CSV file
df = pd.read_csv("telecom_customer_pca_500.csv")

# Display first 5 rows
df.head()
```

```python
# Check dataset shape
print("Dataset shape:", df.shape)
```

```python
# Check column information
df.info()
```

```python
# Display summary statistics
df.describe()
```

## 3. Check Missing Values

```python
# Check missing values
df.isnull().sum()
```

## 4. Select Features for PCA

PCA works on numeric input columns.

We remove `customer_id` because it is only an identifier and does not help in PCA.


```python
# Remove customer_id
X = df.drop("customer_id", axis=1)

# Display selected features
X.head()
```

## 5. Scale the Data

Scaling is very important for PCA.

PCA is affected by the size/range of values.

For example:

- total_charges may be very large
- customer_service_calls may be very small

So we use StandardScaler before applying PCA.


```python
# Create StandardScaler object
scaler = StandardScaler()

# Scale the input features
X_scaled = scaler.fit_transform(X)

print("Scaling completed")
```

## 6. Apply PCA with 2 Components

Here we reduce all numeric columns into only **2 principal components**.

This is useful for visualization.


```python
# Create PCA object with 2 components
pca = PCA(n_components=2)

# Apply PCA
X_pca = pca.fit_transform(X_scaled)

# Convert PCA result into DataFrame
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])

# Display first 5 rows
pca_df.head()
```

## 7. Explained Variance Ratio

Explained variance tells us how much information is captured by each principal component.

Higher explained variance means that component is more useful.


```python
# Explained variance ratio
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

# Total variance captured by PC1 and PC2
print("Total Variance Captured:", pca.explained_variance_ratio_.sum())
```

## 8. Visualize PCA Output

This plot shows customers in a 2D PCA space.

Each point represents one telecom customer.


```python
plt.figure(figsize=(8, 5))
plt.scatter(pca_df["PC1"], pca_df["PC2"], alpha=0.6)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of Telecom Customers")
plt.grid(True)
plt.show()
```

## 9. Find Best Number of Components

Instead of manually selecting 2 components, we can check cumulative explained variance.

Usually, we select enough components to capture around **90% to 95%** of the information.


```python
# Apply PCA without fixing number of components
pca_full = PCA()
pca_full.fit(X_scaled)

# Cumulative explained variance
cumulative_variance = pca_full.explained_variance_ratio_.cumsum()

# Display values
for i, value in enumerate(cumulative_variance, start=1):
    print("Components:", i, "Cumulative Variance:", value)
```

```python
# Plot cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker="o")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

## 10. PCA with 95% Variance

Here we ask PCA to automatically keep enough components to preserve 95% of the information.


```python
# PCA to preserve 95% variance
pca_95 = PCA(n_components=0.95)

X_pca_95 = pca_95.fit_transform(X_scaled)

print("Original number of features:", X.shape[1])
print("Reduced number of components:", X_pca_95.shape[1])
print("Variance captured:", pca_95.explained_variance_ratio_.sum())
```

## 11. Create Final PCA DataFrame

```python
# Create column names dynamically
component_names = [f"PC{i+1}" for i in range(X_pca_95.shape[1])]

# Convert PCA output to DataFrame
pca_95_df = pd.DataFrame(X_pca_95, columns=component_names)

# Add customer_id back for reference
pca_95_df.insert(0, "customer_id", df["customer_id"])

# Display output
pca_95_df.head()
```

## 12. Understand PCA Components

This table shows how original columns contribute to principal components.

Large positive or negative values mean that feature has strong influence on that component.


```python
# PCA component loading table
loadings = pd.DataFrame(
    pca_95.components_,
    columns=X.columns,
    index=component_names
)

loadings
```

## 13. Save PCA Output

```python
# Save PCA-transformed data
pca_95_df.to_csv("telecom_customer_pca_output.csv", index=False)

print("PCA output file saved successfully")
```

## Simple Explanation

PCA does not predict anything.

It is not classification and it is not regression.

It is used to reduce many columns into fewer meaningful components.

For example:

Original telecom data may have 9 numeric columns.

After PCA, it may become 6 or 7 components while still preserving most of the information.

This reduced data can later be used for:

- K-Means clustering
- customer segmentation
- visualization
- machine learning model input
